```python
import numpy as np
import matplotlib.pyplot as plt  # Don't forget to import matplotlib

x = np.linspace(0, 10, 100)
fig, ax = plt.subplots()  # Create a figure and axis
ax.plot(x, np.sin(x), "-")  # Solid line
ax.plot(x, np.sin(x), "--")  # Dashed line

plt.show()
```
